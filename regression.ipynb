```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
X = np.array([0, 21.5, 24.5, 25.5])
y = np.array([0.04, 0.069, 0.112, 0.160])
```

```python
def exponential_func(x, a, b):
    return a * np.exp(b * x)
```

```python
# Fit the exponential curve
params, covariance = curve_fit(exponential_func, X, y)
```

```python
# Get the parameters
a, b = params
print(f"Fitted exponential function: y = {a} * exp({b} * x)")
```

```python
# Predict y values using the fitted function
y_pred = exponential_func(X, a, b)
```

```python
# Calculate R-squared value
r_squared = r2_score(y, y_pred)
print(f"R-squared: {r_squared}")
```

```python
# Plot the data points and the fitted curve
plt.scatter(X, y, label='Data Points')
x_fit = np.linspace(min(X), max(X), 100)
y_fit = exponential_func(x_fit, a, b)
plt.plot(x_fit, y_fit, label=f'Fitted Exponential Curve\ny = {a:.2f} * exp({b:.2f} * x)\nR-squared = {r_squared:.2f}', color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Polynomial regression

```python
# Data
X = np.array([0, 21.5, 24.5, 25.5]).reshape(-1, 1)
y = np.array([0.04, 0.086, 0.138, 0.190])

# Polynomial regression
degree = 2
poly = PolynomialFeatures(degree)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

# Predictions
y_pred = model.predict(X_poly)

# R-squared value
r_squared = r2_score(y, y_pred)
print(f"R-squared: {r_squared}")

# Plotting
plt.scatter(X, y, label='Data Points')
x_fit = np.linspace(min(X), max(X), 100).reshape(-1, 1)
x_fit_poly = poly.transform(x_fit)
y_fit = model.predict(x_fit_poly)
plt.plot(x_fit, y_fit, label=f'Polynomial Degree {degree}\nR-squared = {r_squared:.2f}', color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Logarithmic regression

```python
# Logarithmic function
def logarithmic_func(x, a, b):
    return a * np.log(x) + b

# Fit the logarithmic curve
params, covariance = curve_fit(logarithmic_func, X.flatten(), y)

# Get the parameters
a, b = params
print(f"Fitted logarithmic function: y = {a} * log(x) + {b}")

# Predictions
y_pred = logarithmic_func(X.flatten(), a, b)

# R-squared value
r_squared = r2_score(y, y_pred)
print(f"R-squared: {r_squared}")

# Plotting
plt.scatter(X, y, label='Data Points')
x_fit = np.linspace(min(X), max(X), 100)
y_fit = logarithmic_func(x_fit, a, b)
plt.plot(x_fit, y_fit, label=f'Logarithmic\nR-squared = {r_squared:.2f}', color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

## power law regression

```python
# Power law function
def power_law_func(x, a, b):
    return a * np.power(x, b)

# Fit the power law curve
params, covariance = curve_fit(power_law_func, X.flatten(), y)

# Get the parameters
a, b = params
print(f"Fitted power law function: y = {a} * x^{b}")

# Predictions
y_pred = power_law_func(X.flatten(), a, b)

# R-squared value
r_squared = r2_score(y, y_pred)
print(f"R-squared: {r_squared}")

# Plotting
plt.scatter(X, y, label='Data Points')
x_fit = np.linspace(min(X), max(X), 100)
y_fit = power_law_func(x_fit, a, b)
plt.plot(x_fit, y_fit, label=f'Power Law\nR-squared = {r_squared:.2f}', color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
